**Data visualization**

TERRORIST ATTACKS EACH YEAR

```python
att = gt['DATE'].value_counts(dropna = False).sort_index().to_frame().reset_index().rename(columns = {"index" : "DATE", "DATE" : "ATTACK"}).set_index("DATE")
```

```python
att.head()
```

```python
gt['DATE'] = pd.to_datetime(gt['DATE'])
```

```python
gt['YEAR'] = gt['DATE'].dt.year
print(gt)
```

```python
ywa = gt['YEAR'].value_counts(dropna = False).sort_index().to_frame().reset_index().rename(columns = {"index" : "YEAR", "YEAR" : "ATTACK"}).set_index('YEAR')
```

```python
ywa.plot(kind = "bar", color = "Orange", figsize = (15,6), fontsize = 12)
plt.title("Terrorist Activities", fontsize = 15)
plt.xlabel("YEAR", fontsize = 10)
plt.ylabel("Attacks occurred", fontsize = 10)
plt.show()
```

OBSERVATION
- Terrorist activities were on peak in 2006.
- Least amount of terrorist activities were observed in 2023.
- In early 2000s, the amount of terrorist activities across the globe were on a rise.


CONTINENTAL TERRORIST ACTIVITIES

```python
reg = pd.crosstab(gt.YEAR, gt.REGION)
reg.head()
```

```python
reg.plot(kind = "area", stacked = False , alpha = 0.5 , figsize = (20, 10))
plt.title("Continental terrorist Activities", fontsize = 15)
plt.xlabel("YEAR", fontsize = 12)
plt.ylabel("No. of Attacks", fontsize = 12)
plt.show()
```

OBSERVATION

- As the graph shows, the amount of terrorist activities is tracked in Asia.
- Least amount of activities were recorded in Oceania from the beginning to this year.
- Europe and America is more or less had the same amount of terrorist activities all these years.

NUMBER OF DEATHS OCCURRED

```python
df = pd.DataFrame(gt, columns = ['YEAR', 'DEAD'])
```

```python
deaths = df.groupby('YEAR')['DEAD'].sum()
deaths.plot(kind = 'bar' , color = 'red', figsize = (18, 10), fontsize = 12)
plt.title("Number of Deaths in Each Year", fontsize = 15)
plt.xlabel("YEAR", fontsize = 10)
plt.ylabel("No. of Deaths", fontsize = 10)
plt.show()
```

OBSERVATION

- In 2001, over 3000 people died due to these activities all across the world.
- After 2001, almost 2500 people died in 2016.
- Least amount of death occurred due to terrorism in 1968 and after that in 1970 and 1977.

No. of injured

```python
df2 = pd.DataFrame(gt, columns = ['YEAR', 'INJURED'])
injured = df2.groupby('YEAR')['INJURED'].sum()
injured.plot(kind = 'bar', color = 'green', figsize = (18,10), fontsize = 12)
plt.title("No. of Injured each Year", fontsize = 15)
plt.xlabel("YEAR", fontsize = 10)
plt.ylabel("No. of Injured", fontsize = 10)
plt.show()
```

OBSERVATION
- In 1998, over 5000 people got injured all across the globe due to terrorism.
- In 2005, 2008 and 2016 over 4000 people had serious injuries.
- Least amount of people gott injured in 1968 and 1969.

DISTRIBUTION OF INCIDENTS BY CATEGORY

```python
category = gt.CATEGORY.str.split(',', expand=True).stack().value_counts()
plt.figure(figsize=(8, 8))
plt.pie(category, labels=category.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Distribution of Incidents by Category')
plt.show()
```

OBSERVATION
- Explosions have been predominantly employed (nearly 70%) as a primary means for propagating acts of terrorism.
- On the other hand, occurrence of accidents remains comparatively minimal.

TOP PERPETRATOR INVOLVED IN INCIDENTS

```python
perpetrator = gt.PERPETRATOR.str.split(',', expand = True).stack().value_counts()
top_perpetrator = perpetrator.head(10)
plt.figure(figsize = (8,8))
plt.pie(top_perpetrator, labels = top_perpetrator.index , autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Top Perpetrators Involved in Incidents', fontsize = 15, pad = 20)
plt.show()
```

OBSERVATION
- The provided pie chart illustrates the involvement of various entities in acts of terrorism, with the top 10 perpetrators being highlighted.

- Notably, the Revolutionary Armed Forces of Colombia (FARC) as the foremost contributor (nearly 15%) to these terroristic activities.
- Conversely, the pie chart positions Hezbollah at the lower end of the spectrum, indicating its relatively lesser involvement compared to the other entities listed.
